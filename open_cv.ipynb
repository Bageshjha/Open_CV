```python
#image Transformation
img = cv.imread(r"C:\Users\bages\OneDrive\Pictures\mustang.jpg")
cv.imshow('image',img)

def translate(img , x ,y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1], img.shape[0])#width,height
    return cv.warpAffine(img , transMat , dimensions)


# ( x) -> right
# (-x) -> left
# ( y) -> down
# (-y) -> up

translated = translate(img , 100 ,100)
cv.imshow("translated"  , translated)


#rotating image

def rotate(img, angle , rotPoint = None):
    (height , width) = img.shape[:2]
    
    if rotPoint is None :
        rotPoint = (width // 2 , height //200)
        
    rotMat = cv.getRotationMatrix2D(rotPoint , angle , 1.0)    
    dimensions = (width , height )
    
    return cv.warpAffine(img , rotMat , dimensions)

rotated = rotate(img , 45)
cv.imshow("rotated" , rotated)


#flip
flip = cv.flip(img , 0)
cv.imshow("flip" , flip)

cv.waitKey(0)
```

```python
#bitwise operation
import cv2 as cv
import numpy as np

blank = np.zeros((400,400), dtype = 'uint8')

rectangle = cv.rectangle(blank.copy(), (30,30),(370,370), 255 , -1)

cv.imshow("rectangle",rectangle)

circle = cv.circle(blank.copy(), (200,200) ,200 , 186 ,-1)
cv.imshow("circle", circle)

#bitwise_and
bitwise_and = cv.bitwise_and(rectangle, circle)
cv.imshow("bitwise_and" , bitwise_and)

#bitwise_or
bitwise_or = cv.bitwise_or(rectangle, circle)
cv.imshow("bitwise_or" , bitwise_or)

#bitwise_xor
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv.imshow("bitwise_xor" , bitwise_xor)

#bitwise_not
bitwise_not = cv.bitwise_not(rectangle)
cv.imshow("bitwise_not" , bitwise_not)

cv.waitKey(0)
```

```python
#thresholding

import cv2 as cv
import numpy as np

img = cv.imread(r"C:\Users\bages\OneDrive\Pictures\mustang.jpg")
cv.imshow('image',img)

#simple thresholding

threshold, thresh = cv.threshold(img , 225 , 255 , cv.THRESH_BINARY)
cv.imshow("simple threshold" , thresh)

#simple threshold_inverse
threshold, thresh_inv = cv.threshold(img , 225 , 255 , cv.THRESH_BINARY_INV)
cv.imshow("inverse threshold" , thresh_inv)

#adaptive threshold(works on grayscale only)
gray = cv.cvtColor(img , cv.COLOR_BGR2GRAY)
cv.imshow("gray" , gray)
adaptive_thresh = cv.adaptiveThreshold(gray , 255 , cv.ADAPTIVE_THRESH_MEAN_C , cv.THRESH_BINARY, 11,3)
cv.imshow("adaptive threshold" , adaptive_thresh)

#adaptive threshold_inverse
adaptive_thresh_inv = cv.adaptiveThreshold(gray , 255 , cv.ADAPTIVE_THRESH_MEAN_C , cv.THRESH_BINARY_INV, 11,3)
cv.imshow("adaptive threshold_inv" , adaptive_thresh_inv)

cv.waitKey(0)
```

```python
import cv2 as cv
import numpy as np

img = cv.imread(r"C:\Users\bages\OneDrive\Pictures\mustang.jpg")
cv.imshow('image',img)

#edge detection

gray = cv.cvtColor(img , cv.COLOR_BGR2GRAY)
cv.imshow("gray" , gray)

#laplaction
lap = cv.Laplacian(gray,cv.CV_64F)
lap = np.uint8(np.absolute(lap))
cv.imshow("laplaction" , lap)


#sobel
sobelx = cv.Sobel(gray, cv.CV_64F , 1, 0)
sobely = cv.Sobel(gray, cv.CV_64F , 0, 1)
combined_sobel = cv.bitwise_and(sobelx , sobely)
combine_sobel = cv.bitwise_or(sobelx , sobely)

cv.imshow("sobel_x" , sobelx)
cv.imshow("sobel_y" , sobely)
cv.imshow("combined_sobel" , combined_sobel)
cv.imshow("combine_sobel" , combine_sobel)


cv.waitKey(0)
```
